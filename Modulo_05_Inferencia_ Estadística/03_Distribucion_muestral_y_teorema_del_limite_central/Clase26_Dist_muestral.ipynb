```python
import math
from scipy import stats

# 1. Datos del problema
mu = 15          # Media poblacional
sigma = 5       # Desviación estándar
n = 36          # Tamaño de la muestra
x_bar = 16      # Media muestral a evaluar

# 2. Cálculo del Error Estándar
error_estandar = sigma / math.sqrt(n)

# 3. Cálculo del puntaje Z
z = (x_bar - mu) / error_estandar

# 4. Cálculo de la probabilidad P(X > 16)
# stats.norm.sf es la "Survival Function" (1 - CDF), que da el área a la derecha
probabilidad = stats.norm.sf(z)

print(f"Puntaje Z: {z:.2f}")
print(f"Probabilidad: {probabilidad:.4f} ({probabilidad*100:.2f}%)")
```

```python
import math
from scipy import stats

# 1. Datos registrados
m1, s1, n1 = 130, 10, 30  # Grupo Control
m2, s2, n2 = 125, 12, 30  # Grupo Tratado

# 2. Diferencia de medias
delta = m1 - m2

# 3. Error estándar para diferencias
error_estandar = math.sqrt((s1**2 / n1) + (s2**2 / n2))

# 4. Cálculo de Z y P-valor
z_score = delta / error_estandar
# Multiplicamos por 2 para una prueba de dos colas (cambio en cualquier sentido)
p_valor = stats.norm.sf(abs(z_score)) * 2

# 5. Tamaño del efecto (d de Cohen para dos muestras)
# Usamos la desviación estándar combinada (pooled)
s_pooled = math.sqrt(((s1**2) + (s2**2)) / 2)
d_cohen = delta / s_pooled

print(f"Error Estándar: {error_estandar:.4f}")
print(f"Valor Z: {z_score:.4f}")
print(f"P-valor: {p_valor:.4f}")
print(f"d de Cohen: {d_cohen:.4f}")
```
