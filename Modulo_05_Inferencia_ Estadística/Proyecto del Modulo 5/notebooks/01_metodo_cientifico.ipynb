# Método Científico y Diseño de Investigación Estadística

## Contexto

Los servicios de salud universitarios desean comprender los factores que influyen en los hábitos saludables de los estudiantes.

Estos hábitos incluyen la **calidad del sueño**, la **actividad física** y los **patrones de alimentación**.

Comprender estos comportamientos puede ayudar a diseñar mejores programas de bienestar estudiantil.

## Problema de Investigación

¿Qué factores influyen en los hábitos de vida saludables entre los estudiantes universitarios?

Específicamente, analizamos las relaciones entre las **horas de sueño**, la **actividad física**, la **alimentación** y los **niveles de estrés**.

## Hipótesis

**Hipótesis Nula (H0):**

No existe una relación significativa entre los hábitos saludables (sueño y ejercicio) y los niveles de estrés entre los estudiantes universitarios.

**Hipótesis Alternativa (H1):**

Los estudiantes con mejores hábitos saludables (más horas de sueño y mayor actividad física) tienden a presentar niveles de estrés más bajos.

## Variables

**Variables cuantitativas:**

- Edad
- Horas de sueño por noche
- Horas de ejercicio por semana
- Comidas de comida rápida por semana
- Nivel de estrés (escala de 1 a 10)
- Promedio académico (GPA)

**Variables cualitativas:**

- Género
- Nivel de consumo de alcohol

## Diseño del Estudio

El estudio utilizará un conjunto de datos **simulado** que representa a estudiantes universitarios.

**Pasos:**

1. Simular un conjunto de datos de al menos 100 estudiantes
2. Explorar modelos de probabilidad
3. Analizar distribuciones estadísticas
4. Aplicar el Teorema del Límite Central
5. Estimar intervalos de confianza
6. Realizar pruebas de hipótesis
