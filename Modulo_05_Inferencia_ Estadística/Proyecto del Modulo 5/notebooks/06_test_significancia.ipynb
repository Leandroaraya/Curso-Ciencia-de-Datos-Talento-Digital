# Pruebas de Hipótesis y Test de Significancia

## Objetivo

Validar afirmaciones sobre la población de estudiantes mediante pruebas de hipótesis para medias y proporciones.

Se calculará el valor-p y se comparará con el nivel de significancia α para decidir si se rechaza o no la hipótesis nula.

## Variables de interés

- Horas de sueño
- Nivel de estrés
- Dormir bien (≥7 horas, variable binaria)

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv("/content/drive/MyDrive/habitos_saludables_estudiantes.csv")
data

# Crear variable binaria dormir bien
data["duerme_bien"] = data["horas_sueno"] >= 7
```

Ejemplo 1: Test de media (una muestra)

Supongamos que la media poblacional de horas de sueño esperada es 7 horas y queremos ver si nuestros estudiantes difieren.

Hipótesis:

H0: μ = 7

H1: μ ≠ 7

```python
sueno = data["horas_sueno"]
mu_hipotetico = 7
alpha = 0.05

t_stat, p_val = stats.ttest_1samp(sueno, mu_hipotetico)
print(f"T-statistic: {t_stat:.3f}, p-valor: {p_val:.3f}")

if p_val < alpha:
    print("Rechazamos H0: La media de horas de sueño es significativamente diferente de 7")
else:
    print("No rechazamos H0: No hay evidencia para decir que la media difiere de 7")
```

Ejemplo 2: Test de proporciones

Supongamos que queremos ver si al menos el 50% de los estudiantes duermen bien (≥7 horas).

Hipótesis:

H0: p = 0.5

H1: p ≠ 0.5

```python
duerme_bien = data["duerme_bien"]
n = len(duerme_bien)
p_obs = duerme_bien.mean()
p0 = 0.5

# Test de proporción usando aproximación normal
z = (p_obs - p0) / np.sqrt(p0*(1-p0)/n)
p_val = 2*(1 - stats.norm.cdf(abs(z)))

print(f"Proporción observada: {p_obs:.3f}, z: {z:.3f}, p-valor: {p_val:.3f}")

if p_val < alpha:
    print("Rechazamos H0: La proporción de estudiantes que duermen bien difiere de 50%")
else:
    print("No rechazamos H0: No hay evidencia para decir que difiere del 50%")
```

## Errores en pruebas de hipótesis

- **Error Tipo I (α)**: Rechazar H0 siendo verdadera.  
  En nuestro caso, afirmar que la media de horas de sueño difiere de 7 cuando realmente es 7.

- **Error Tipo II (β)**: No rechazar H0 siendo falsa.  
  En nuestro caso, no detectar que la media difiere de 7 cuando en realidad sí difiere.

El valor de α se establece antes del test (0.05 en este ejemplo) y controla la probabilidad de cometer un error tipo I.
