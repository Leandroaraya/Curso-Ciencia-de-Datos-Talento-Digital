```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
```

Cargar datos>>Limpiarlos>>Separar los datos X,y>>Escalarlos>>Crear Modelos

```python
data = fetch_california_housing()
data
```

```python
print(data.feature_names)
print(data.data)
print('-------------------------')
print(data.target_names)
print(data.target)


```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
print(df.head())
```

```python
df['MedHouseVal'] = data.target
display(df.head())
```

```python
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']
display(X)
display(y)
print(len(X))
```

```python
#normalizar las variables
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
len(X_scaled)
display(X_scaled)
```

```python
display(X_scaled[0])
```

```python
#Separar la data Entrenamiento y Test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
print(X_train.shape) #entrenamiento con el 80%
print(X_test.shape) #test con el 20% que corresponde al dato indicado en test_size=0.2
```

#crear modelos de regresion lineal

https://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_scaled,y) # r2
```

```python
model.coef_ #coeficientes del modelo: y=b0+b1*X1+b2*X2+...+b8*X8
```

```python
b0=model.intercept_ # b0
b0
```

```python
# Validacion cruzada k-fold (dividiendo los datos de entrenemientos y validacion en subconjuntos  mas pequeños)
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,shuffle=True,random_state=43)
kf
```

```python
#obtenemos los de r2 para cada fold (para cada una de las vueltas determina un r2 con los sub conjuntos de test del split)
scores=cross_val_score(model,X_train,y_train,cv=kf,scoring='r2')
scores
```

```python
#ver estabilidad del modelo
promedio_score=np.mean(scores)
desviacion_score=np.std(scores)
print(promedio_score)
print(desviacion_score)
```
