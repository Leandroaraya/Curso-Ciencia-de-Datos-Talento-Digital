# Redes Neuronales Artificiales

Requieren alto volumenes de datos para garantizar su efectividad
pueden resolver problemas  complejos

Funcion de activacion: es la funcion que se evalua con respecto a la sumatoria de los pesos, considerando el sesgo.
y=f(xi*wi+b): x:variable entrada,w:peso,b:sesgo
funcion de perdida como se cuantifica el error del valor real con el predicho
suelen usarse:MSE(para regresion),Cross-Entropy(para clasificacion)

Ciclo de entrenamiento:
forward pass: datos de entradas pasan por respectivos pesos iniciales y luego a la funcion de activacion. se compara la salida esperada con la obtenida y se calcula el error 
backpropagation: del error vuelve hacia atras a los pesos para luego ser ajustados. este proceso se repite multiples epocas.

Capa entrada-Capa de preprocesamiento- Capa de salida (Capa de entrada,Capa Oculta..Capa Oculta,Capa de salida)
variables de entradas->parametros(pesos)

Hiperparametros de entrenamiento:
Learning rate:controla los pasos del descenso de gradiente 
Batch-Size: cantidad por iteracion del descenso de gradiente
Epochs:numero de veces que el algoritmo procesa el conjunto de datos de entrenamiento

existen optimizadores mas avanzados que el descendo de gradiente como:
Adam,RMSprop,Adagrad,Momentum

Regularizacion para evitar overfitting:
Dropout:apaga aleatoriamente algunas neuronas
L1/L2: penaliza la funcion de perdida
Early Stopping:detiene el entrenamiento cuando el conjunto de validacion comientza a empeorar

La evalaucion de los modelos es comun dividirlas en 3 conjuntos
Entrenamiento:para ajustar los pesos de la red
Validacion: Para ajustar hiperparametros y ver sobreajuste
Prueba:evaluar el rendimiento final del modelo

Metricas de evaluacion:
clasificacion:Accuracy(precision)  precision(Exactitud)   Recall(Sensibilidad)    F1-Score   AUC-ROC
Regresion: Mean squared Error (MSE)   RMSE     Mean Absolute Error (MAE)      R2(Coeficiente determinacion)   

OTROS TIPOS DE REDES NEURONALES
* Redes Neuronales Convolucionales CNN -> imagenes o estructura datos especiales
* Redes Neuronales Recurrentes RNN     -> secuencias y datos temporales
* Redes Generativas Adversarias GAN    -> para generar datos similares a los de entrenamiento
