# Determinar cuantos años y dias me faltan para jubilar

```python
#Determinar cuantos años y dias me faltan para jubilar

año_jubilacionx=65;
año_nacimientox=date(1992,3,13);
empleadox="Leandro"
Edadx=date.today()-año_nacimientox;#esto me da la diferencia en dias
print(Edadx)

fecha_jubilacionx=date(año_nacimientox.year+año_jubilacionx,año_nacimientox.month,año_nacimientox.day)
print(fecha_jubilacionx)


# Diferencia inicial
años = fecha_jubilacionx.year - date.today().year
meses = fecha_jubilacionx.month - date.today().month
dias = fecha_jubilacionx.day - date.today().day

print(años)
print(meses)
print(dias)

```

```python
###Mejorar por problema al restar fechas

# Ajustes
if dias < 0:
    meses -= 1
    # días del mes anterior
    mes_anterior = date.today().month - 1 if date.today().month > 1 else 12
    año_mes_anterior = date.today().year if date.today().month > 1 else date.today().year - 1
    
    dias_mes_anterior = (date(año_mes_anterior, mes_anterior + 1, 1) - date(año_mes_anterior, mes_anterior, 1)).days
    dias += dias_mes_anterior

if meses < 0:
    años -= 1
    meses += 12

print(f"Te faltan {años} años, {meses} meses y {dias} días para jubilar 🎉")

"""
Este código resuelve el problema de ajustar correctamente la diferencia entre dos fechas cuando los días o meses resultan negativos.

Cuando se calcula la diferencia entre fechas (años, meses y días), puede ocurrir que:

Los días sean negativos (por ejemplo, si el día actual es mayor que el día de la fecha objetivo)
O que los meses sean negativos

Para corregir esto, el código:

Si los días son negativos → “pide prestado” un mes y suma los días del mes anterior
Si los meses son negativos → “pide prestado” un año y ajusta los meses

De esta forma, se obtiene un resultado correcto y entendible como:
“X años, Y meses y Z días” en lugar de valores negativos.

📅 Hoy: 15 de octubre de 2026
🎯 Jubilación: 10 de marzo de 2057

años  = 2057 - 2026 = 31
meses = 3 - 10 = -7  ❌
dias  = 10 - 15 = -5 ❌
"""
```
