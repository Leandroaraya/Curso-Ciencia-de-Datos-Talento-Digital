```python
# 1. Pedir datos al usuario
nombre = input("Ingrese su nombre: ")
edad = input("Ingrese su edad: ")

nota1 = input("Ingrese la primera nota: ")
nota2 = input("Ingrese la segunda nota: ")
nota3 = input("Ingrese la tercera nota: ")

# 4. Conversiones de tipo (aquí está lo que te piden)
edad = int(edad)              # str → int
nota1 = float(nota1)          # str → float
nota2 = float(nota2)
nota3 = float(nota3)

# 2.a Promedio
promedio = (nota1 + nota2 + nota3) / 3

# 2.b Edad en meses
edad_meses = edad * 12

# 2.c Evaluar si aprueba (comparación lógica)
aprueba = promedio >= 6

# 3. Mostrar resultados
print("\n--- RESULTADOS ---")
print(f"Hola {nombre}")
print(f"Tu edad en meses es: {edad_meses}")
print(f"Tu promedio es: {promedio:.2f}")

if aprueba:
    print("Estado: Aprobado ✅")
else:
    print("Estado: Reprobado ❌")
```

```python
# usando funcion para pedir notas y que sea numerico

def pedir_nota(mensaje):
    while True:
        try:
            return float(input(mensaje))
        except ValueError:
            print("Error: ingresa un número válido")

nota1 = pedir_nota("Nota 1: ")
nota2 = pedir_nota("Nota 2: ")
nota3 = pedir_nota("Nota 3: ")

promedio = (nota1 + nota2 + nota3) / 3
print(f"Tu promedio es: {promedio:.2f}")
```
