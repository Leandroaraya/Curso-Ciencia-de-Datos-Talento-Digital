## Programación Orientada a Objeto

```python
class Usuario:
    def __init__(self, nombre, email, password):
        self.nombre = nombre
        self.email = email
        self.password = password

    def verificar_password(self, password):
        return self.password == password
```

```python
class SistemaLogin:
    def __init__(self):
        self.usuarios = []

    def registrar_usuario(self, usuario):
        self.usuarios.append(usuario)
        print(f"Usuario {usuario.nombre} registrado correctamente")

    def login(self, email, password):
        for usuario in self.usuarios:
            if usuario.email == email:
                if usuario.verificar_password(password):
                    print(f"Bienvenido {usuario.nombre}")
                    return True
                else:
                    print("Contraseña incorrecta")
                    return False
        print("Usuario no encontrado")
        return False
```

```python
# Crear usuarios
usuario1 = Usuario("Leandro", "lea@email.com", "1234")
usuario2 = Usuario("Ana", "ana@email.com", "abcd")
```

```python
sistema = SistemaLogin()

sistema.registrar_usuario(usuario1)
sistema.registrar_usuario(usuario2)
```

```python
sistema.login("lea@email.com", "1234")   # correcto
sistema.login("lea@email.com", "0000")   # contraseña incorrecta
sistema.login("otro@email.com", "1234")  # usuario no existe
```

```python
email = input("Ingrese su email: ")
password = input("Ingrese su contraseña: ")

sistema.login(email, password)
```
