BLOQUE 1: Definir la clase Empleado

```python
class Empleado:
    def __init__(self, nombre, rol, salario):
        self.nombre = nombre
        self.rol = rol
        self.salario = salario

    def mostrar(self):
        print(f"Nombre: {self.nombre} | Rol: {self.rol} | Salario: {self.salario}")

    def actualizar_salario(self, nuevo_monto):
        self.salario = nuevo_monto

    def gana_mas_que(self, promedio):
        return self.salario > promedio
```

BLOQUE 2: Crear empleados

```python
empleado1 = Empleado("Leandro", "Ingeniero", 1200000)
empleado2 = Empleado("Ana", "Analista", 900000)
empleado3 = Empleado("Pedro", "Supervisor", 1100000)
```

BLOQUE 3: Guardarlos en una lista

```python
empleados = [empleado1, empleado2, empleado3]
```

BLOQUE 4: Mostrar información

```python
print("📋 Lista de empleados:\n")

for emp in empleados:
    emp.mostrar()
```

BLOQUE 5: Calcular promedio de salarios

```python
suma_salarios = 0

for emp in empleados:
    suma_salarios += emp.salario

promedio = suma_salarios / len(empleados)

print(f"\n💰 Promedio de salarios: {promedio}")
```

BLOQUE 6: Ver quién gana más que el promedio

```python
print("\n🚀 Empleados que ganan más que el promedio:\n")

for emp in empleados:
    if emp.gana_mas_que(promedio):
        print(emp.nombre)
```

```python
empleado2.actualizar_salario(1300000)

print("\n🔄 Después de actualizar salario:\n")

for emp in empleados:
    emp.mostrar()
```
