## Sentencias Iterativas While-For-Range


```python
numeros=[1,2,3,4,5,6,3,2,4,52,1]
lista_pares=[]
print("numeros pares de la lista")
for i in numeros:
    if i%2==0:
        lista_pares.append(i)
        print(i)
print(lista_pares)
```

```python
#aplicar un contador
contador=0
while contador<=5:
    print("contador: ",contador)
    contador+=1
   
```

```python
x=5
y=5
while x==y:
    x+=1
    print("se finalizó")
    break #para salir del codigo si no esta x+=1
```

```python
#contar frutas
frutas=["manzana","piña","frutilla","sandia","melon","platano","manzana","frutilla","manzana"]

from collections import Counter

conteo = Counter(frutas)

print(conteo)
```

```python
frutas = ["manzana","piña","frutilla","sandia","melon","platano","manzana","frutilla","manzana"]

conteo = {}

for fruta in frutas:
    if fruta in conteo:
        conteo[fruta] += 1
    else:
        conteo[fruta] = 1 #aqui se inicia conteo de la primera aparicion  porque conteo es vacio

print(conteo)
```

```python
numeros = [1, 2, 3, 4]

# Forma tradicional
nueva_lista = []
for n in numeros:
    nueva_lista.append(n)

# List comprehension otra forma
nueva_lista2 = [n for n in numeros]
nueva_lista3 = [n*2 for n in numeros]
nueva_lista4 = [n*2 for n in numeros if n >2]#primero filtra [3,4] luego multiplica


print(nueva_lista)
print(nueva_lista2)
print(nueva_lista3)
print(nueva_lista4)
```

```python
#for en diccionarios
personas={"Ana":22,"juan":30}

for nombre,edad in personas.items(): #con items permite tomar ambos(clave y valor) si no lo coloco tomara por defecto la clave
    print(nombre,edad)
#otra forma
print("----------------------")
for nombre in personas.keys():
    print(nombre)
print("----------------------")
for nombre in personas.values():
    print(nombre)
```

```python
h=list(range(11))
print(h) # forma simple recorre hasta el 11 sin considerarlo
print(list(range(2,11))) # inicia del 2
print(list(range(2,11,2))) # hace saltos de 2 en 2

for x in range(2,7,2):
    print(x)

for i in range(len(h)): # aqui sumamos 5 a cada valor
    h[i]+=5
print(h)
```

```python
#para enumerar
frutas = ["manzana","piña","frutilla","sandia","melon","platano","manzana","frutilla","manzana"]
for indice,i in enumerate(frutas):
    print(indice,i)
```

```python
#uso del zip
nombres=["Leandro","Felipe","Mena"]
edades=[32,33,34]
for nombre,edad in zip(nombres,edades):
    print(nombre,edad)
```
