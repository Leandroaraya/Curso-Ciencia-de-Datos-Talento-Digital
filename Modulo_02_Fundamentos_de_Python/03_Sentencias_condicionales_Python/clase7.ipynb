```python
# Entradas
monto = float(input("Ingrese el monto del carrito: "))
medio_pago = input("Medio de pago (efectivo/tarjeta): ").lower()
es_estudiante = input("¿Es estudiante? (s/n): ").lower()
tiene_cupon = input("¿Tiene cupón? (s/n): ").lower()
envio_rapido = input("¿Envío rápido? (s/n): ").lower()

# -------------------------
# DESCUENTO BASE
# -------------------------
if monto >= 80000:
    descuento = 0.15
elif monto >= 50000:
    descuento = 0.10
elif monto >= 30000:
    descuento = 0.05
else:
    descuento = 0.0

# -------------------------
# EXTRA ESTUDIANTE
# -------------------------
if es_estudiante == "s" and medio_pago == "tarjeta":
    descuento += 0.05

# -------------------------
# CUPÓN
# -------------------------
if tiene_cupon == "s" and (es_estudiante != "s" or monto < 50000):
    descuento += 0.05

# -------------------------
# TOPE DESCUENTO
# -------------------------
if descuento > 0.25:
    descuento = 0.25

# -------------------------
# CÁLCULO TOTAL CON DESCUENTO
# -------------------------
total_descuento = monto * (1 - descuento)

# -------------------------
# ENVÍO
# -------------------------
envio = 2000  # estándar

# condición de envío gratis
envio_gratis = (monto >= 60000) or (monto >= 40000 and es_estudiante == "s")

if envio_gratis:
    envio = 0

# si quiere envío rápido
if envio_rapido == "s":
    if envio_gratis:
        envio = 4000
    else:
        envio = 6000

# -------------------------
# TOTAL FINAL
# -------------------------
total_final = total_descuento + envio

# -------------------------
# MENSAJE TERNARIO
# -------------------------
financiacion = "APLICA FINANCIACIÓN" if (medio_pago == "tarjeta" and total_final >= 70000) else "SIN FINANCIACIÓN"

# -------------------------
# SALIDA
# -------------------------
print("\n------ TICKET FINAL ------")
print(f"Subtotal: ${monto:,.0f}")
print(f"Descuento aplicado: {descuento*100:.0f}%")
print(f"Total con descuento: ${total_descuento:,.0f}")
print(f"Envío: ${envio:,.0f}")
print(f"TOTAL FINAL: ${total_final:,.0f}")
print(financiacion)
```
