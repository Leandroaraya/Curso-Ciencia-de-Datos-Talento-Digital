#  Estructura de datos en Python

```python
#creamos una lista
lista=[1,2,3,'texto',4.5]
print(type(lista))
lista.append(5) #agrega al final de la lista 
print(lista)
lista.insert(1,"a") #agrega señalando la ubicacion en la posicion 1 colocare esto y de alli se desplaza a la derecha 
print(lista)
print(lista[0])
print(lista[1:])# desde la 1 hasta el final
print(lista[:4])#  considera desde el principio hasta posicion 4-1
print(lista[1:4]) #desde la posicion 1 hasta la 4 sin incluirla, osea hasta la 3
```

```python
matriz=[[1,2,3],[4,5,6],[7,8,9]]
print(matriz)
print(matriz[1][1])
print(matriz[0][1:4]) #matriz[0][1:]
```

```python
lista.remove("a") #solo acepta un argumento el remove
lista.remove("texto")
```

```python

print(lista)
lista.append(0)
print(lista)
lista.sort()#ordena la lista de menor a mayor
print(lista)
lista.sort(reverse=True)#ordena la lista de mayor a menor
print(lista)

```

```python
lista.pop()#elimina el ultimo elemento
print(lista)
```

```python
lista.pop(1)#elimina por posicion
print(lista)
```

```python
del lista[0] #elimina por posicion
print(lista)
del lista[1:3] #elimina de la posicion 1 hasta la 3-1
print(lista)
```

```python
sorted(lista)
print(lista)#no cambia la lista. 
print(sorted(lista))
lista1=[1,2,3]
lista2=[4,5,6]
lista1.extend(lista2)
print(lista1)
print(lista1.index(5))#devuelve la posicion del numero 5, osea la posicion 4
```

```python
#diccionarios
diccionario={"nombre":"Leandro","apellido":"Araya","Edad":33,"telefono":934118335}
print(diccionario)
print(diccionario["nombre"])#se obtienen los valores (clave-valor)
diccionario["nombre"]="Felipe"
print(diccionario["nombre"])
type(diccionario)
```

```python
#rellenar una lista 
personas = {
    1: {"nombre": "Leandro", "apellido1": "Araya", "apellido2": "Mena", "nota1": 6.0, "nota2": 5.5, "nota3": 6.5},
    2: {"nombre": "Juan", "apellido1": "Rosas", "apellido2": "Mena", "nota1": 5.0, "nota2": 4.5, "nota3": 6.0},
    3: {"nombre": "Ana", "apellido1": "Perez", "apellido2": "Soto", "nota1": 6.2, "nota2": 5.8, "nota3": 6.1},
    4: {"nombre": "Luis", "apellido1": "Gomez", "apellido2": "Diaz", "nota1": 4.0, "nota2": 5.0, "nota3": 4.5},
    5: {"nombre": "Carla", "apellido1": "Muñoz", "apellido2": "Rojas", "nota1": 6.8, "nota2": 6.5, "nota3": 6.9},
    6: {"nombre": "Pedro", "apellido1": "Torres", "apellido2": "Vega", "nota1": 5.5, "nota2": 5.0, "nota3": 5.2},
    7: {"nombre": "Sofia", "apellido1": "Castro", "apellido2": "Lagos", "nota1": 6.0, "nota2": 6.1, "nota3": 6.3},
    8: {"nombre": "Diego", "apellido1": "Herrera", "apellido2": "Silva", "nota1": 4.8, "nota2": 5.2, "nota3": 5.0},
    9: {"nombre": "Valentina", "apellido1": "Flores", "apellido2": "Nuñez", "nota1": 6.7, "nota2": 6.4, "nota3": 6.6},
    10: {"nombre": "Martin", "apellido1": "Reyes", "apellido2": "Cortes", "nota1": 5.9, "nota2": 5.7, "nota3": 6.0}
}
lista_nombres = []

for persona in personas.values():
    nombre_completo = f"{persona['nombre']} {persona['apellido1']} {persona['apellido2']}"
    lista_nombres.append(nombre_completo)

print(lista_nombres)
print(list(map(lambda x:x,lista_nombres)))

for nombre in lista_nombres:
    print(nombre)
```

```python
#crear una tupla
tupla=(1,2,3,3,4)# se usan () y es inmutable a diferencia de la lista
#crear un set
set1={0,10,2,1,4,4,4,2,9,6,7} # el set no permite valores repetidos. los eliminara y reordenara
print(tupla,set1)
set1.add(8) #agrega un dato adicional
set1.discard(1) #para eliminar un dato
print(set1)
setnuevo=set(tupla)
print(setnuevo,"sssssssssssss")# igual se puede transformar tupla con valor repetido a set. se eliminan los repetidos
listax=list(tupla) # para modificar una tupla hay qye cambiarla a lista
listax.remove(3)
print(listax)
transforma=set(listax)
print(transforma)
```
