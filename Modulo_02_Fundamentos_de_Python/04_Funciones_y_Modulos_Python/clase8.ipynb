```python
#funcion que puedo llamarla sin atributos pero queda preestablecido su atributo
def saludo(receptor="estimado"):
    print(f"Hola {receptor}")
```

```python
def saludin(receptor="amig@"):
    saludo_final=f"hola  {receptor}"
    return saludo_final
```

```python
#INDICE MASA CORPORTAL
def calcular_IMC(peso,altura):
    return peso/altura**2
```

```python
def clasificar_IMC(IMC):
    if IMC<18.5:
        resultado="bajo de peso"
    elif IMC>=18.5 and IMC<=24.9:
        resultado="peso normal"
    else:
        resultado="sobrepeso"
    return resultado

```

```python
imc=calcular_IMC(85,1.75)
print(imc)
clasificar_IMC(imc)
```
