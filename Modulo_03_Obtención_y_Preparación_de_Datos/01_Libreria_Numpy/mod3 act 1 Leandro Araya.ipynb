1. Carga y estructuración de datos:

○ Crea un array NumPy con datos financieros simulados (por ejemplo, precios de acciones en diferentes días).

○ Organiza los datos en una matriz de 5x5, donde cada fi la representa una acción y cada columna un día de cotización.

```python
import numpy as np

# Simulación de precios de 5 acciones durante 5 días
precios = np.array([
    [100, 102, 101, 103, 105],
    [50, 49, 51, 52, 50],
    [200, 198, 202, 205, 207],
    [80, 82, 81, 83, 85],
    [120, 121, 119, 122, 124]
])

print("Matriz de precios:")
print(precios)
```

```python
#caso sin numpy:
# Simulación de precios de 5 acciones durante 5 días
precioss = [
    [100, 102, 101, 103, 105],
    [50, 49, 51, 52, 50],
    [200, 198, 202, 205, 207],
    [80, 82, 81, 83, 85],
    [120, 121, 119, 122, 124]
]

print("Matriz de precios:")
for fila in precioss:
    print(fila)
```

2. Análisis y transformación de datos:

○ Obtén el promedio, valor máximo y mínimo de cada acción a lo largo del tiempo.

○ Calcula la variación porcentual diaria de cada acción.

○ Aplica funciones matemáticas como logaritmo, exponencial o normalización sobre los datos.

```python
#Obtén el promedio, valor máximo y mínimo de cada acción a lo largo del tiempo.
promedio = np.mean(precios, axis=1) # axis =1 operar por filas, axis=0 operar por columnas
maximo = np.max(precios, axis=1)
minimo = np.min(precios, axis=1)
#axis=1 operar por filas,Toma cada fila completa: Calcula el promedio / máximo / mínimo de esa fila
print("Promedio por acción:", promedio)
print("Máximo por acción:", maximo)
print("Mínimo por acción:", minimo)
```

```python
#sin numpy hay que aplicar mas codigos y anidarlos
promedios = []
maximos = []
minimos = []

for accion in precioss:
    suma = 0
    max_val = accion[0]
    min_val = accion[0]

    for precio in accion:
        suma += precio
        if precio > max_val:
            max_val = precio
        if precio < min_val:
            min_val = precio

    promedios.append(suma / len(accion))
    maximos.append(max_val)
    minimos.append(min_val)

print("Promedio por acción:", promedios)
print("Máximo por acción:", maximos)
print("Mínimo por acción:", minimos)
```

```python
#Calcula la variación porcentual diaria de cada acción.
variacion_porcentual = np.diff(precios, axis=1) / precios[:, :-1] * 100

print("Variación porcentual diaria:")
print(variacion_porcentual)
```

```python
#sin numpy
variacion_porcentual = []

for accion in precioss:
    variacion_accion = []
    for i in range(1, len(accion)):
        cambio = (accion[i] - accion[i - 1]) / accion[i - 1] * 100
        variacion_accion.append(cambio)
    variacion_porcentual.append(variacion_accion)

print("Variación porcentual diaria:")
for v in variacion_porcentual:
    print(v)
```

```python
#Aplica funciones matemáticas como logaritmo, exponencial o normalización sobre los datos.
log_precios = np.log(precios)
exp_precios = np.exp(precios / 100)
normalizados = (precios - precios.min()) / (precios.max() - precios.min())
print("Logaritmo de los precios:")
print(log_precios)
print("Exponencial de los precios:")
print(exp_precios)
print("Precios normalizados:")
print(normalizados)
```

```python
import math

#sin numpy
log_precios = []

for accion in precioss:
    log_accion = []
    for precio in accion:
        log_accion.append(math.log(precio))
    log_precios.append(log_accion)

print("Logaritmo de precios:")
for fila in log_precios:
    print(fila)





exp_precios = []

for accion in precioss:
    exp_accion = []
    for precio in accion:
        exp_accion.append(math.exp(precio / 100))
    exp_precios.append(exp_accion)


# Buscar mínimo y máximo global
min_global = precios[0][0]
max_global = precios[0][0]

for accion in precios:
    for precio in accion:
        if precio < min_global:
            min_global = precio
        if precio > max_global:
            max_global = precio



# Normalización
normalizados = []

for accion in precioss:
    norm_accion = []
    for precio in accion:
        norm = (precio - min_global) / (max_global - min_global)
        norm_accion.append(norm)
    normalizados.append(norm_accion)

print("Precios normalizados:")
for fila in normalizados:
    print(fila)
```
