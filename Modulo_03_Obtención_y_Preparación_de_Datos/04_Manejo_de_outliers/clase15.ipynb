# Manejo de outliers

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'oldName': [1, 2, 3], 'anotherOldName': ['a', 'b', 'c']}) 

#renombrar columnas de un DataFrame
print(df)
print("------------------   ------------------")
df.rename(columns={'oldName': 'newName', 'anotherOldName': 'anotherNewName'}, inplace=True)
print(df)
print("------------------   ------------------")
  
#cambiar valores de una columna
df["anotherNewName"] = df["anotherNewName"].replace({'a': 'x','b': 'y','c': 'z'}) 
print(df)
print("------------------   ------------------")
#cambiar valores de una columna
df["anotherNewName"] = df["anotherNewName"].replace('x', 'nuevoValor') 
print(df)
print("------------------   ------------------")
#cambiar valores de una columna
df["anotherNewName"].replace('nuevoValor', 'asdfghjkl', inplace=True) 
print(df)
```

```python
df.isnull().sum() #nulos por columna  
(df.isnull().sum() / len(df)) * 100 #porcentaje de nulos por columna  
df.isnull().sum().sum() #nulos totales
df.dropna(thresh=2) #elimina filas: thresh=2 → mínimo 2 datos válidos.Menos de eso → se elimina
```

```python
df = pd.DataFrame({
    "Nombre": [
        "Ana", "Luis", "María", "Pedro", "Sofía",
        "Jorge", "Camila", "Diego", "Valentina", "Carlos",
        "Fernanda", "Andrés", "Paula", "Ricardo", "Daniela",
        "Tomás", "Claudia", "Felipe", "Antonia", "Miguel"
    ],
    "Area trabajo": [
        "IT", "Ventas", "IT", "Marketing", "Ventas",
        "IT", "RRHH", "Ventas", "Marketing", "IT",
        "RRHH", "Ventas", "IT", "Marketing", "Ventas",
        "IT", "RRHH", "Marketing", "Ventas", "IT"
    ],
    "Salario": [
        1200, 6000, 1300, 1100, 1050,
        None, 1150, 980, None, 1250,
        1400, None, 1350, 1120, 1080,
        None, 1500, 1180, 1020, None
    ]
})
print(df)
```

```python
df.iloc[5,0] 
```

```python
df.loc[5, 'Area trabajo'] = np.nan
print(df.head(10))
print("------------------   ------------------")
df.iloc[8,1] = np.nan
df.iloc[8,0] = np.nan
print(df.head(10))
```

```python
df.dropna(thresh=1,inplace=True) ##elimina filas: thresh=1 → mínimo 1 datos válidos.menos de eso → se elimina
#aqui se fue el dato 8 porque tenia solo un valor en la fila.
#df.dropna(thresh=2) # thresh=2 → eliminaria dos filas: la 5 y la 8 
print(df)
```

```python
#df.dropna()#elimina todas las filas que tengan valores NaN
#df.dropna(axis=0)
df.dropna(axis=1) #elimina columnas con valores NaN
```

```python
df.groupby("Area trabajo")["Salario"].mean().sort_values(ascending=False)
```

```python
df["Salario"] = df["Salario"].fillna(
    df.groupby("Area trabajo")["Salario"].transform("mean")
)
df
```

```python
print(df)
```

```python
#cuartiles
Q1 = df["Salario"].quantile(0.25)
Q2 = df["Salario"].quantile(0.50) #mediana
Q3 = df["Salario"].quantile(0.75)
print(f"Q1: {Q1}, Q2: {Q2}, Q3: {Q3}")
IQR = Q3 - Q1
print(f"IQR: {IQR}")
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Limite inferior: {lower_bound}, Limite superior: {upper_bound}")
print("cualquier valor fuera de estos límites es un outlier")
print(df[(df["Salario"] < lower_bound) | (df["Salario"] > upper_bound)])
print("-------------------   ------------------")
print(df[(df["Salario"] >= lower_bound) & (df["Salario"] <= upper_bound)])
```
