```python
#las series tienen un índice asociado a cada elemento, que por defecto es un número entero que comienza en 0
serie=pd.Series([100, 200, 300, 400, 500])
print(serie)
print(serie[0])
#serie indezada a partir de uma lista de índices
serie2=pd.Series([100, 200, 300, 400, 500], index=['a', 'b', 'c', 'd', 'e'])
print(serie2)
print(serie2['b'])
print(serie2.iloc[3]) #acceso por posición posicion 0,1,2,3,4
print(serie2.loc["a"]) #acceso por etiqueta o index
```

```python
print(df3["Edad"].min()) #minimo de la columna "Edad"
print(df3["Edad"].max()) #maximo de la columna "Edad"
print(df3["Edad"].mean()) #media de la columna "Edad"
print(df3["Edad"].median()) #mediana de la columna "Edad"
print(df3["Edad"].std()) #desviación estándar de la columna "Edad"
print(df3["Edad"].count()) #número de valores de la columna "Edad"
print(df3["Edad"].sum()) #suma de los valores de la columna "Edad"
print("---------------")

valoresUnicos=df3["Ciudad"].value_counts() #conteo de valores únicos en la columna "Ciudad"
print(valoresUnicos)
print("---------------")
print("frecuencia relativa (%):",df3["Ciudad"].value_counts(normalize=True) * 100) #frecuencia relativa de valores únicos en la columna "Ciudad"
print("---------------")
print("valores únicos:",df3["Ciudad"].unique()) #valores únicos en la columna "Ciudad"
print("---------------")
print("numero de valores únicos:",df3["Ciudad"].nunique()) #número de valores únicos en la columna "Ciudad"
```
