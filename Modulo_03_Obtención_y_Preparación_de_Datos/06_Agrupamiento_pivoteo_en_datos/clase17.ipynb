### Agrupamiento y pivoteo de datos

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.DataFrame({
    "Departamento": [
        "Ventas","Ventas","TI","TI","Marketing",
        "RRHH","Finanzas","Ventas","TI","Marketing",
        "RRHH","Finanzas","Ventas","Ventas","TI",
        "Marketing","RRHH","Finanzas","Ventas","TI",
        "Ventas","Marketing","RRHH","Finanzas","Ventas"
    ],
    "Region": [
        "Norte","Sur","Norte","Sur","Norte",
        "Sur","Norte","Sur","Norte","Sur",
        "Sur","Norte","Sur","Norte","Sur",
        "Norte","Sur","Norte","Sur","Norte",
        "Sur","Norte","Sur","Norte","Sur"
    ],

    "Empleado": [
        "Juan","María","Pedro","Ana","Luis",
        "Sofía","Carlos","Diego","Valentina","Camila",
        "Andrés","Fernanda","Jorge","Paula","Ricardo",
        "Daniela","Tomás","Claudia","Felipe","Antonia",
        "Miguel","Lucía","Matías","Gabriela","Hugo"
    ],
    "Ventas": [
        50000,60000,70000,80000,55000,
        52000,75000,62000,72000,58000,
        53000,77000,61000,61000,71000,
        59000,54000,78000,61000,73000,
        63000,60000,55000,79000,62000
    ]
})

print(df)
```

```python
#agrupar pot departamentos y calcular el salario promedio por departamento
salario_promedio = df.groupby("Departamento")["Ventas"].mean()
print("\n Ventas promedio por departamento:")
print(salario_promedio)
```

```python
#Ventas por region
Ventas_region_departamento=df.groupby(["Region", "Departamento"])["Ventas"].sum()
print("\n Ventas por region y departamento:")  
print(Ventas_region_departamento)
```

```python
#agrupar pot departamentos y calcular el salario promedio por departamento y sumar las ventas por departamento
salario_promedio = df.groupby("Departamento")["Ventas"].agg(["mean", "sum"])
print("\n Ventas promedio por departamento:")
print(salario_promedio)
```

```python
salario_promedio_y_total = df.groupby("Departamento").agg({"Ventas": ["mean", "sum"]})
print("\n Ventas promedio y total por departamento:")
print(salario_promedio_y_total)
```

```python
despivotear = salario_promedio.stack().reset_index()
print(despivotear)
```

```python
df_pivot= df.pivot_table(values="Ventas", index="Departamento", columns="Region", aggfunc=["sum","mean","count"])
print("\n Tabla pivote de ventas por departamento y region:")   
print(df_pivot)
```

```python
df_pivot= df.pivot_table(values="Ventas", index="Departamento", columns="Region")
print("\n Tabla pivote de ventas por departamento y region:")   
print(df_pivot)
print("-----------------------------")
#melt
despivotear = df_pivot.reset_index().melt(id_vars="Departamento", var_name="Region", value_name="Valor")
print("\n Tabla despivotada:")
print(despivotear)
```

```python
#ejemplo merge
#1. Dataset de empleados
df_empleados = pd.DataFrame({
    "id_empleado": [101,102,103,104],
    "nombre": ["Juan","María","Pedro","Ana"],
    "departamento": ["Ventas","TI","Marketing","RRHH"]
})

#2. Dataset de salarios
df_salarios = pd.DataFrame({
    "id_empleado": [101,102,103,104],
    "salario": [50000,60000,55000,52000]
})

#3 Data set de nuevos ingresos
df_nuevos= pd.DataFrame({
    "id_empleado": [105,106],
    "nombre": ["Luis","Sofía"],
    "departamento": ["Finanzas","Ventas"]
})

print("\n Dataset de empleados:")
print(df_empleados)
print("\n Dataset de salarios:")
print(df_salarios)
print("\n Dataset de nuevos ingresos:")
print(df_nuevos)
```

```python
nuevo=pd.merge(df_empleados, df_salarios, on="id_empleado", how="inner")
print("\n Dataset combinado (inner join):")
print(nuevo)
```

```python
nuevo1=df_empleados.merge(df_salarios, on="id_empleado", how="inner")
print("\n Dataset combinado (inner join):")
print(nuevo)
```
