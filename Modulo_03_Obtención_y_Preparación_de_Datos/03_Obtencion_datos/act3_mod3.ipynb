2. Limpieza y estructuración de datos:
○ Identifica valores nulos y decide si deben ser imputados o eliminados.

○ Verifica y ajusta los tipos de datos en columnas numéricas y categóricas.

```python
#detectar valores nulos
df_excel.isnull().sum()
```

```python
df_excel = df_excel.dropna(subset=["Ventas"])
df_excel.isnull().sum()
```

```python
#   Ver duplicados exactos
df_excel.duplicated().sum()
```

```python
#Revisar tipos de datos actuales
df_excel.dtypes
#se vizualiza que los datos estan correctamente asignados
```

3. Transformación y optimización de datos:
○ Selecciona las columnas más relevantes para el análisis.
○ Renombra columnas para mejorar la legibilidad.
○ Ordena los datos en función de una columna clave.

```python
df_excel = df_excel[['Nombre', 'Ventas']]
df_excel
```

```python
#ordenar datos en base una columna
df_excel = df_excel.sort_values(by='Ventas', ascending=False)
df_excel
```

#4. Conclusiones sobre la importancia de un proceso estructurado de obtención de datos.

Un proceso estructurado de obtención de datos es fundamental para asegurar la calidad, confiabilidad y utilidad de la información analizada. La correcta recopilación, limpieza, transformación y validación de los datos permite reducir errores, evitar inconsistencias y facilitar el análisis posterior. Además, trabajar con datos bien organizados mejora la toma de decisiones, optimiza el tiempo de trabajo y garantiza resultados más precisos y reproducibles. En este contexto, contar con un flujo ordenado desde la obtención hasta la exportación de los datos es clave para cualquier proyecto de análisis de datos.
