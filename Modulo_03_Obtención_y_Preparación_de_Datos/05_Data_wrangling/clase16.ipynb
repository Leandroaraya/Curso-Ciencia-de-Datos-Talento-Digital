```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Nombre": [
        "Ana", "Luis", "María", "Pedro", "Sofía",
        "Jorge", "Camila", "Diego", "Valentina", "Carlos",
        "Fernanda", "Andrés", "Paula", "Ricardo", "Daniela",
        "Tomás", "Claudia", "Felipe", "Antonia", "Miguel"
    ],
    "Area trabajo": [
        "IT", "Ventas", "IT", "Marketing", "Ventas",
        "IT", "RRHH", "Ventas", "Marketing", "IT",
        "RRHH", "Ventas", "IT", "Marketing", "Ventas",
        "IT", "RRHH", "Marketing", "Ventas", "IT"
    ],
    "Salario": [
        1200, 6000, 1300, 1100, 1050,
        None, 1150, 980, None, 1250,
        1400, None, 1350, 1120, 1080,
        None, 1500, 1180, 1020, None
    ]
})
print(df)
```

```python
#reordenar las filas
df_reordenado = df.sample(frac=1).reset_index(drop=True) #frac=1 para tomar todas las filas, reset_index(drop=True) para resetear el índice sin agregar una nueva columna
df_reordenado2= df.sample(frac=0.5).reset_index(drop=True) #frac=0.5 para tomar la mitad de las filas
print(df_reordenado)
print(df_reordenado2)
df_reordenado3=df.sample(n=3).reset_index(drop=True) #n=3 para tomar 3 filas aleatorias
print(df_reordenado3)   
```

```python
df_reordenado=df_reordenado.sort_values(by="Salario", ascending=False).reset_index(drop=True) #ordenar por salario de mayor a menor
print(df_reordenado)    
```

```python
df_reordenado=df_reordenado.sort_values(by=["Salario","Area trabajo"], ascending=[False,True]).reset_index(drop=True) #ordenar por salario de mayor a menor
print(df_reordenado)
```

```python
print(df.duplicated().sum()) #contar filas duplicadas
df_sin_duplicados=df.drop_duplicates()#eliminar filas duplicadas
df.reset_index(drop=True, inplace=True) #resetear el índice sin agregar una nueva columna y modificar el DataFrame original
```

```python
mapa = {
    "IT": "Tecnología",
    "Ventas": "Sales",
    "Marketing": "MKT",
    "RRHH": "Recursos Humanos"
}

df["Area trabajo"] = df["Area trabajo"].map(mapa)
print(df)
```

```python
#categorizar una fila salario
def categorizar_salario(salario):
    if salario < 1000:
        return "Bajo"
    elif salario < 1200:
        return "Medio"
    else:
        return "Alto"
    
#aplicar la función a la columna salario
df["Categoria salario"] = df["Salario"].apply(categorizar_salario)
print(df)
```

```python
df["Salario_bono"] = df["Salario"].apply(lambda x: x * 1.1)
print(df)
df["Salario_bono"] = df["Salario"]*0.9
print(df)
```

```python
#cambiar tipo de dato de la columna salario a int
#df["Salario"] = df["Salario"].astype(int) #si hay NaN produce error
df["Salario"] = df["Salario"].astype("Int64")
print(df.info())
```

```python
#definir una columna como indice
df.set_index("Nombre", inplace=True)
print(df)
```
